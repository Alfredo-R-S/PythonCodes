# 2. Find another identifier variable in the df_ords dataframe that doesn’t need to be included in your analysis as a numeric variable and change it to a suitable format.

```python
df_ords.head()
```

```python
# First, the column eval_set will be removed from the original dataset as it is no longer necessary.
df_ords = df_ords.drop(columns = ['eval_set'])
```

```python
df_ords.head()
```

```python
df_ords.describe()
```

```python
# Both "order_id" and "user_id" do not represent real numeric values but rather keys related to specific orders and users.
# For this reason, their data type will be changed from numeric to string.
df_ords['order_id'] = df_ords['order_id'].astype('str')
df_ords['user_id'] = df_ords['user_id'].astype('str')
```

```python
# Both "order_id" and "user_id" are not included into the statistical analysis anymore.
df_ords.describe()
```

# 3. Look for a variable in your df_ords dataframe with an unintuitive name and change its name without overwriting the dataframe.

```python
# The column order_dow wil be changed into orders_day_of_week but the false argument will be used in inplace =, so we dont overwrite the original column name
df_ords.rename(columns = {'order_dow' : 'orders_day_of_week'}, inplace = False)
```
