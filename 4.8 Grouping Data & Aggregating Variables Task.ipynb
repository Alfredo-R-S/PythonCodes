# 4. Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
# A new column is created called “max_order,” into which the results of the aggregation will be placed.
ords_prods_merge_2['max_order'] = ords_prods_merge_2.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# With the new 'max_order' column ready, a 'flag' column will be created to assign a 'loyalty' label to a user ID based on its corresponding max order value.
ords_prods_merge_2.loc[ords_prods_merge_2['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge_2.loc[(ords_prods_merge_2['max_order'] <= 40) & (ords_prods_merge_2['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge_2.loc[ords_prods_merge_2['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# The value_counts() function will check the values in the new column
ords_prods_merge_2['loyalty_flag'].value_counts(dropna = False)
```

```python
# A general overview of the DataFrame is performed to check for irregularities in the new column.
ords_prods_merge_2.head()
```

# 6. The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
# If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
# If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# A new column is created called “max_price,” into which the results of the aggregation will be placed.
ords_prods_merge_2['mean_price'] = ords_prods_merge_2.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# With the new 'mean_price' column ready, a 'flag' column will be created to classify a user ID's spending type based on its corresponding mean_price value
ords_prods_merge_2.loc[ords_prods_merge_2['mean_price'] >= 10, 'spending_type'] = 'High spender'
ords_prods_merge_2.loc[ords_prods_merge_2['mean_price'] < 10, 'spending_type'] = 'Low spender'
```

```python
# The value_counts() function will check the values in the new column 'spending_type'
ords_prods_merge_2['spending_type'].value_counts(dropna = False)
```

```python
# A general overview of the DataFrame is performed to check for irregularities in the new column.
ords_prods_merge_2.head()
```

# 7. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
# If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
# If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
# If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# A new column is created called “max_price,” into which the results of the aggregation will be placed.
ords_prods_merge_2['median_frequency'] = ords_prods_merge_2.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# With the new 'median_frequency' column ready, a 'flag' column will be created to classify users
# as frequent or non-frequent customers based on their corresponding median days_since_prior_order values.
ords_prods_merge_2.loc[ords_prods_merge_2['median_frequency'] > 20, 'frequent_customers'] = 'Non-frequent customer'
ords_prods_merge_2.loc[(ords_prods_merge_2['median_frequency'] <= 20) & (ords_prods_merge_2['median_frequency'] > 10), 'frequent_customers'] = 'Regular customer'
ords_prods_merge_2.loc[ords_prods_merge_2['median_frequency'] <= 10, 'frequent_customers'] = 'Frequent customer'
```

```python
# The value_counts() function will check the values in the new column 'frequent_customers'
ords_prods_merge_2['spending_type'].value_counts(dropna = False)
```
