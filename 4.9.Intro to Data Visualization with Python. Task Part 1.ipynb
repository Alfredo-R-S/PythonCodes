# 5. Complete the fundamental data quality and consistency checks you’ve learned throughout this Achievement; for example, check for and address missing values and duplicates, and convert any mixed-type data.

```python
# It is possible to check for any unusual values in the columns. Based on this analysis, there do not appear to be any particularly strange numbers.
df_cust.describe()
```

```python
# Here it is possible to observe whether there are any missing values in the DataFrame.
df_cust.isnull().sum()
```

```python
# Here it is possible to observe whether there are any mixed-type data in the DataFrame.
for col in df_cust.columns.tolist():
  weird = (df_cust[[col]].map(type) != df_cust[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_cust[weird]) > 0:
    print (col)
```

```python
# The column First Name contains mix-type data, It will be change into a String type
df_cust['First Name'] = df_cust['First Name'].astype('str')
```

```python
# A check is perfomred, showing that the column First Name no longer contais mixed-type data
for col in df_cust.columns.tolist():
  weird = (df_cust[[col]].map(type) != df_cust[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_cust[weird]) > 0:
    print (col)
```

```python
# A second check, shows no more missing values beacuase the names with NaN were change into a String type data.
df_cust.isnull().sum()
```

```python
# Here a check for duplicates was performed
df_cust_dup = df_cust[df_cust.duplicated()]
```
