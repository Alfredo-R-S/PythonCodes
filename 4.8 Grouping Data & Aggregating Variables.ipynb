```python
# Import libraries
import pandas as pd
import numpy as np
import os
```

```python
# The ords_prods_merge DataFrame is imported
df_ords_prods_merge_2 = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_merge_2.pkl'))
```

```python
# Only 1,000,000 rows will be imported 
df = df_ords_prods_merge_2[:1000000]
```

```python
# A general overview of the DataFrame is performed to check for irregularities.
df.head()
```

```python
# this function will group a given dataframe by a given column
df.groupby('product_name')
```

```python
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
df.groupby('department_id')['order_number'].mean()
```

```python
df.groupby('department_id').order_number.mean()
```

```python
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

```python
df_ords_prods_merge_2['max_order'] = df_ords_prods_merge_2.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merge_2.loc[df_ords_prods_merge_2['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df_ords_prods_merge_2.loc[(df_ords_prods_merge_2['max_order'] <= 40) & (df_ords_prods_merge_2['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df_ords_prods_merge_2.loc[df_ords_prods_merge_2['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```
